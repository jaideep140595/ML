```python
import matplotlib.pyplot as plt
import numpy as np
import math
import cv2
```

```python
imgr = cv2.imread('1.jpg',0)
imgg = cv2.imread('2.jpg',0)
imgb = cv2.imread('3.jpg',0)
imgi = cv2.imread('4.jpg',0)
```

```python
X_riv=[]
Y_riv=[]
X_nriv=[]
Y_nriv=[]
```

```python
def click_event(event, x, y, flags, params):
    if event == cv2.EVENT_LBUTTONDOWN:
        X_riv.append(x)
        Y_riv.append(y)
        print(len(X_riv))
cv2.imshow('image', imgr)
cv2.setMouseCallback('image', click_event)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
def click_event(event, x, y, flags, params):
    if event == cv2.EVENT_LBUTTONDOWN:
        X_nriv.append(x)
        Y_nriv.append(y)
        print(len(X_nriv))
cv2.imshow('image', imgr)
cv2.setMouseCallback('image', click_event)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
train_riv=[]
train_nriv=[]
for i in X_riv:
    for j in Y_riv:
        t=[]
        t.append(imgr[i,j])
        t.append(imgg[i,j])
        t.append(imgb[i,j])
        t.append(imgi[i,j])
    train_riv.append(t)
for i in X_nriv:
    for j in Y_nriv:
        t=[]
        t.append(imgr[i,j])
        t.append(imgg[i,j])
        t.append(imgb[i,j])
        t.append(imgi[i,j])
    train_nriv.append(t)
```

```python
mean_riv=[]
mean_nriv=[]
r=0
g=0
b=0
t=0
for i in train_riv:
    r+=i[0]
    g+=i[1]
    b+=i[2]
    t+=i[3]
mean_riv.append(r/len(train_riv))
mean_riv.append(g/len(train_riv))
mean_riv.append(b/len(train_riv))
mean_riv.append(t/len(train_riv))
r=0
g=0
b=0
t=0
for i in train_nriv:
    r+=i[0]
    g+=i[1]
    b+=i[2]
    t+=i[3]
mean_nriv.append(r/len(train_nriv))
mean_nriv.append(g/len(train_nriv))
mean_nriv.append(b/len(train_nriv))
mean_nriv.append(t/len(train_nriv))
```

```python
cov_riv=np.cov(np.transpose(train_riv))
cov_nriv=np.cov(np.transpose(train_nriv))
print(cov_riv)
```

```python
test_data=np.zeros((512,512,4))
for i in range(512):
    for j in range(512):
        test_data[i][j][0]=imgr[i,j]
        test_data[i][j][1]=imgg[i,j]
        test_data[i][j][2]=imgb[i,j]
        test_data[i][j][3]=imgi[i,j]
test_data.shape
```

```python
mean_riv=np.array(mean_riv)
sub_m=[]
for i in range(512):
    t=[]
    for j in range(512):
        t.append(list(test_data[i][j]-mean_riv))
    sub_m.append(t)
sub_m=np.array(sub_m)
print(sub_m.shape)
```

```python
Out_image1 = [[0 for i in range(512)] for j in range(512)]
Out_image2 = [[0 for i in range(512)] for j in range(512)]
Out_image3 = [[0 for i in range(512)] for j in range(512)]
```

```python
a = []
riv_class = 0
b = []
non_riv_class = 0
M1 = 0.3
N1 = 0.7
M2 = 0.7
N2 = 0.3
M3 = 0.5
N3 = 0.5
for i in range(512):
    for j in range(512):
        a = np.subtract(test_data[i][j], mean_riv)
        riv_class = np.dot((np.dot(a, np.linalg.inv(cov_riv))),np.transpose(a))
        p1 = 1/(math.sqrt(np.linalg.det(cov_riv)))*np.exp(riv_class*(-0.5))
        b = np.subtract(test_data[i][j], mean_nriv)
        non_riv_class = np.dot((np.dot(b, np.linalg.inv(cov_nriv))),np.transpose(b))
        p2 = 1/(math.sqrt(np.linalg.det(cov_nriv)))*np.exp(non_riv_class*(-0.5))
        if(p1*M1 >= p2*N1):
            Out_image1[j][i] = 255
        else:
            Out_image1[j][i] = 0
            
        
        if(p1*M2 >= p2*N2):
            Out_image2[j][i] = 255
        else:
            Out_image2[j][i] = 0
            
        if(p1*M3 >= p2*N3):
            Out_image3[j][i] = 255
        else:
            Out_image3[j][i] = 0
```

```python
plt.imshow(Out_image1)
plt.show()

plt.imshow(Out_image2)
plt.show()

plt.imshow(Out_image3)
plt.show()
```
